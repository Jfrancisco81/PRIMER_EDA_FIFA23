```python


# Obtener el porcentaje de jugadores por pie hbil
porcentaje_PIE_HABIL = fifa23['Pie_habil'].value_counts(normalize=True) * 100

# Definir etiquetas y colores para la grfica
labels = porcentaje_PIE_HABIL.index.tolist()
colors = ['#FFC107', '#2196F3']

# Separar el sector del pie para los jugadores diestros
explode = [0.1, 0]

# Crear la grfica de pie
fig = plt.figure(figsize=(8, 6))
plt.pie(porcentaje_PIE_HABIL, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%')

# Aadir un ttulo a la grfica
plt.title('Porcentaje de jugadores por pie hbil en FIFA 23', fontsize=16, fontweight='bold')

# Mostrar la grfica
plt.show()
```
